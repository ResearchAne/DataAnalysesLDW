```python
V = cv2.GaussianBlur(V,(5,5),cv2.BORDER_DEFAULT)
sobelx = cv2.Sobel(V, cv2.CV_64F,1,0,ksize=1)

plt.imshow(sobelx)
```

```python
abs_sobelx = np.absolute(sobelx)
scaled_sobelx = np.uint8(255*abs_sobelx/np.max(abs_sobelx))
thresh_min = 20
thresh_max = 100
sxbinaryx = np.zeros_like(scaled_sobelx)
sxbinaryx[(scaled_sobelx >= thresh_min) & (scaled_sobelx <= thresh_max)] = 1
plt.axis('off')
plt.imshow(sxbinaryx)
plt.savefig('C:/Users/aneds/Pictures/LD/'+title+'/'+title+'Sobelx'+'.jpg', dpi=300, bbox_inches='tight')
```

```python
sobely = cv2.Sobel(V, cv2.CV_64F,0,1, ksize=1)
abs_sobely = np.absolute(sobely)
scaled_sobely = np.uint8(255*abs_sobely/np.max(abs_sobely))
thresh_min = 30
thresh_max = 100
sxbinaryy = np.zeros_like(scaled_sobely)
sxbinaryy[(scaled_sobely >= thresh_min) & (scaled_sobely <= thresh_max)] = 1
plt.axis('off')
plt.imshow(sxbinaryy)
plt.savefig('C:/Users/aneds/Pictures/LD/'+title+'/'+title+'Sobely'+'.jpg', dpi=300, bbox_inches='tight')
```
